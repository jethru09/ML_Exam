```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-darkgrid')

# Define the domain
x = np.arange(0.5, 100.5, 0.5)

# Define the functions
y1 = x
y2 = x ** 2
y3 = (x**3) /100
y4 = np.sin(x)
y5 = np.sin(x)/ x
y6 = np.log(x)
y7 = np.exp(x)

# Plot the functions
fig, axs = plt.subplots(7, 1, figsize=(8, 32)) # width and height in inches.
axs[0].plot(x, y1)
axs[0].set_title("y = x")
axs[1].plot(x, y2)
axs[1].set_title("y = x^2")
axs[2].plot(x, y3)
axs[2].set_title("y = x^3 /100")
axs[3].plot(x, y4)
axs[3].set_title("y = sin(x)")
axs[4].plot(x, y5)
axs[4].set_title("y = sin(x)/x")
axs[5].plot(x, y6)
axs[5].set_title("y = log(x)")
axs[6].plot(x, y7)
axs[6].set_title("y = e^x")

# Set the x-axis label
fig.text(0.5, 0.04, 'x', ha='center')

# Set the y-axis label
fig.text(0.04, 0.5, 'y', va='center', rotation='vertical')

# Show the plot
plt.show()
```

Note the $$e^x$$ values are represented in the order of 1e43 that is $$1 \times 10^{43}$$ order.
Graphs are ploted seperately.
