```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def __str__(self):
        return f"\n        Name: {self.name}, Roll Number: {self.roll_number}\n            Marks: {self.marks}"


class Branch:
    def __init__(self, name, students):
        self.name = name
        self.students = students

    def __str__(self):
        return f" \n    {self.name}: \n      Students: {', '.join(str(student) for student in self.students)}"


class Year:
    def __init__(self, year, branches):
        self.year = year
        self.branches = branches

    def __str__(self):
        return f"Year: {self.year} \n  Branches: {', '.join(str(branch) for branch in self.branches)}"


# Example:
student1 = Student("John Doe", "21001", {"Math": 90, "English": 85, "Science": 92})
student3 = Student("Jhonny Depp", "21001", {"Math": 88, "English": 90, "Science": 95})
student2 = Student("Jane", "21002", {"Math": 58, "English": 70, "Science": 65})
student4 = Student("Hanna", "21002", {"Math": 58, "English": 70, "Science": 65})
student5 = Student("Peter", "22001", {"Math": 90, "English": 85, "Science": 92})
student6 = Student("Emma", "22001", {"Math": 88, "English": 90, "Science": 95})
student7 = Student("Harry", "22002", {"Math": 58, "English": 70, "Science": 65})
student8 = Student("Ron", "22002", {"Math": 58, "English": 70, "Science": 65})

branch1 = Branch("Computer Science", [student1, student2])
branch2 = Branch("Electrical Engineering", [student3, student4])
branch3 = Branch("Computer Science", [student5, student6])
branch4 = Branch("Electrical Engineering", [student7, student8])

year1 = Year(2021, [branch1, branch2])
year2 = Year(2022, [branch3, branch4])

database = [year1, year2]

# Print the database
for year in database:
    print(year)
```
